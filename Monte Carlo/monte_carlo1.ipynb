```
import math
import random
import matplotlib.pyplot as plt
```

```
n_list = [int(x) for x in input("Input n values: ").split()]    #list(map(int,input().split()))
r = int(input("radius of the circle: "))
x1,y1 = [int(x) for x in input("Center of circle in x y: ").split()]
x_low,x_high = [int(x) for x in input("Input lower and upper bound of x: ").split()]
y_low,y_high = [int(x) for x in input("Input lower and upper bound of y: ").split()]

hit_list, miss_list = [], []
pi_list = []
area_list = []

for n in n_list:
  hit_x,hit_y,miss_x,miss_y = [], [], [], []
  for i in range(0, n):
    x2 = random.uniform(x_low ,x_high)
    y2 = random.uniform(y_low, y_high)
    dist = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    if dist <= r:
      hit_list.append(dist)
      hit_x.append(x2)
      hit_y.append(y2)
    else:
      miss_list.append(dist)
      miss_x.append(x2)
      miss_y.append(y2)
      
  plt.scatter(hit_x, hit_y, color= "green")
  plt.scatter(miss_x, miss_y, color='red')
  plt.show()

  area_rec = len(hit_list)*x_high*2*y_high*2
  pi_list.append(area_rec/(n*r**2))
  area_list.append(area_rec/n*r**2)

  hit_list = []
  miss_list =[]

n_list = [str(i) for i in n_list]
#bar plot for Trials vs. Pi values
plt.bar(n_list,pi_list,color = ['orange','purple'])
plt.xlabel("Number of trials")
plt.ylabel("Values of PI")
plt.title("Fig: Trials Vs. PI graph")
plt.show()

#bar plot for trials Vs. area of circle
plt.bar(n_list,area_list,color = ['orange','purple'])
plt.xlabel("Number of trials")
plt.ylabel("Area of circle")
plt.title("Fig: Trials Vs. Area of circle")
plt.show()
```
