```
import random
import math
import matplotlib.pyplot as plt
```

```
n,d,l = [int(x) for x in input("Input n,d,l values: ").split()]

hit_list= []
theta_hit_list= []

miss_list = []
theta_miss_list= []

for i in range(0, n):
  D = random.uniform(0, d/2)
  theta = random.uniform(0, math.pi)
  
  if (D <= (l/2)*math.sin(theta)):
    hit_list.append(D)
    theta_hit_list.append(theta)
  else:
    miss_list.append(D)
    theta_miss_list.append(theta)

pi = (2*l/d)*(n/len(hit_list))
print("Value of pi: {}".format(pi))

#scatter plot for hit and miss
plt.scatter(theta_hit_list,hit_list,label="Hit",color = "green")
plt.scatter(theta_miss_list,miss_list,label = "miss",color = "red")
plt.xlabel("Theta")
plt.ylabel("D")
plt.title("Fig : Theta vs D graph")
plt.legend(loc = "upper right")
plt.show()
```
